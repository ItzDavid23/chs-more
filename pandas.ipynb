```python
import pandas as pd

serie1 = pd.Series([2, 4, 6, 8, 10])
serie2 = pd.Series([1, 3, 5, 7, 9])

print(f'Suma: \n{serie1 + serie2} \n Resta: \n{serie1 - serie2} \n Multiplicación: \n{serie1 * serie2}\n División: \n{serie1 / serie2}')
```
